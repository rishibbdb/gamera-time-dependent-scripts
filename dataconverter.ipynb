```python
from uncertainties import unumpy
KM2AData = Table()
KM2AData.meta["keywords"] = {"cl": {"value": 0.95}, "name": "KM2A (ApJS 271:25)", "color":"green"}
int_flux = 1.64 * 1e-16 # 1/TeV/cm2/s
int_flux_err = 0.10 * 1e-16 # 1/TeV/cm2/s
spec = 2.75
spec_err = 0.07
E0 = 50 # TeV, 25-340
spec = unumpy.uarray(spec, spec_err)
int_flux = unumpy.uarray(int_flux/TeVtoerg, int_flux_err/TeVtoerg) # 1/erg/cm2/s
pivot = E0 * TeVtoerg # erg
en =  np.logspace(np.log10(25), np.log10(340), 5) * TeVtoerg # erg
sed = int_flux * pow((en/pivot), -spec) * pow(en, 2) # erg/cm2/s
KM2AData["energy"] = Column(en, unit=u.erg, dtype=np.float64)
KM2AData["flux"] = Column(unumpy.nominal_values(sed), unit=u.erg/u.cm**2/u.s, dtype=np.float64)
KM2AData["flux_error_lo"] = Column(unumpy.std_devs(sed), unit=u.erg/u.cm**2/u.s, dtype=np.float64)
KM2AData["flux_error_hi"] = Column(unumpy.std_devs(sed), unit=u.erg/u.cm**2/u.s, dtype=np.float64)
KM2AData["is_ul"] = Column(np.zeros(len(en)), unit='', dtype=np.int64)
KM2AData.write("data/km2aData.csv", format="csv", overwrite=True)
```
